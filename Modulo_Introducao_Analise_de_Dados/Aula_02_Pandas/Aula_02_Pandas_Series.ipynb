# Series
Imagine que você é um pesquisador em uma instituição de pesquisa de conservação da vida selvagem. Você tem uma série de registros sobre os avistamentos de diferentes espécies de animais em uma floresta ao longo de um ano. Esses registros são como pequenos blocos de informações, cada um representando um dia. Agora, você deseja analisar e entender melhor esses dados para tirar conclusões importantes para a preservação da floresta.

No mundo da programação, esses registros podem ser comparados às Series no Pandas. Cada valor na Series representa um dia específico de observação.

# Exploração Inicial

Para começar nossa análise, você decide dar uma olhada nos primeiros registros de avistamentos na floresta. É como abrir o caderno de observações e verificar as primeiras páginas.

No Pandas, você pode criar uma Series com uma semana de informações sobre os avistamentos.

```python
import pandas as pd

avistamentos = pd.Series([10, 15, 8, 20, 12, 18, 9])
print(avistamentos)
```

```python
# Criar índices personalizados
avistamentos = pd.Series([10, 15, 8, 20, 12, 18, 9], index=["Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado", "Domingo"])
print(avistamentos)
```

```python
print(avistamentos['Quarta'])
```

```python
# Acessando os 5 primeiros registros
print(avistamentos.head())
```

```python
# Acessando os 5 últimos registros
print(avistamentos.tail())
```

# Métodos de Estatísticas Descritivas

Na pesquisa, você frequentemente precisa obter informações importantes sobre os dados. Em nosso caso, você deseja saber o número total de avistamentos que ocorreram durante o ano na floresta a média, o desvio padrão, o valor mínimo e o valor máximo. Essas estatísticas ajudarão você a entender melhor os padrões de avistamentos na floresta.

```python
# Calculando o total de avistamentos em uma semana
total_avistamentos = avistamentos.sum()
print(total_avistamentos)
```

```python
# Calculando a média e o desvio padrão de avistamentos em uma semana
media_avistamentos = avistamentos.mean()
std_avistamentos = avistamentos.std()
print(f'A média de avistamentos foi {media_avistamentos:.2f} com o desvio padrão de {std_avistamentos:.2f}')
```

```python
# Calculando o valor mínimo e o valor máximo
valor_minimo = avistamentos.min()
valor_maximo = avistamentos.max()
indice_minimo = avistamentos.index[avistamentos.argmin()]
indice_maximo = avistamentos.index[avistamentos.argmax()]
print(f"Valor mínimo {valor_minimo} no índice {indice_minimo}")
print(f"Valor máximo {valor_maximo} no índice {indice_maximo}")
```

```python
# Observando estatísticas descritivas sobre os dados
print(avistamentos.describe())
```

```python
# Observando estatísticas descritivas sobre os dados
print(f'{avistamentos.describe()[2]:.2f}')
```

# Métodos para Manipulação e Transformação de Dados

À medida que você explora mais os dados, você se depara com um problema: alguns registros de avistamentos estão incompletos. Por exemplo, você não tem dados para os últimos três dias do ano. Para lidar com isso, você decide preencher esses registros com zero

```python
# Importar o NumPy para inserir valores NaN
import numpy as np

avistamentos = pd.Series([10, 15, np.nan, 20, np.nan, 18, 9], index=["Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado", "Domingo"])

print(f'Series sem alteração:  \n{avistamentos}')

avistamentos_completos = avistamentos.fillna(0)
print(f'Series alterada: \n{avistamentos_completos}')
```

```python
print(avistamentos_completos ** 2)
print(np.log(avistamentos_completos))
```

# Seleção

Agora, você está interessado apenas nos registros de avistamentos que excedem um certo limite, digamos, 15. Isso ajuda a identificar dias com observações significativas.

```python
# Selecionando dias com avistamentos acima de 15
dias_significativos = avistamentos[avistamentos > 15]
print(dias_significativos)
```

# Avaliando o Sabor

Como críticos gastronômicos, a avaliação é fundamental. Vamos calcular a média das avaliações dos restaurantes:

# Preparando uma Análise Detalhada

```python
# Filtrando os restaurantes acima da média com preço médio abaixo de R$32,00
restaurantes_acima_media = df[(df["Avaliação"] > media_avaliacoes) & (df["Preço Médio"] < 32)]
print(restaurantes_acima_media)
```

```python
# Filtrando apenas algumas colunas dos restaurantes acima da média
restaurantes_acima_media = df[["Restaurante", "Culinária"]][df["Avaliação"] > media_avaliacoes]
print(restaurantes_acima_media)
```

1. Verifique quais os bairros possuem escolas no conjunto de dados.
2. Descubra o número de alunos por bairro.
3. Descubra o número de escolas por bairro.
4. Descubra o número de alunos aprovados por bairro.
5. Descubra o número de alunos reprovados por frequência que estudam pela MANHÃ em cada escola.
6. Descubra o número de alunos reprovados por frequência que estudam pela NOITE em cada Bairro.

```python
# Ler base de dados
df_escolas = pd.read_csv("http://dados.recife.pe.gov.br/dataset/ce5168d4-d925-48f5-a193-03d4e0f587c7/resource/9afdd9e1-9bc2-4d40-a070-4dec351d6e0f/download/situacaofinalalunos2022.csv", sep=";")
df_escolas.head()
```

```python
# 1. Verifique quais os bairros possuem escolas no conjunto de dados.
bairros_escolas = df_escolas["ebairrnome"].unique()
print(bairros_escolas)
```

```python
# 2. Descubra o número de alunos por bairro.
alunos_por_bairro = df_escolas["ebairrnome"].value_counts()
print(alunos_por_bairro)
```

```python
# 3. Descubra o número de escolas por bairro

escolas_por_bairro = df_escolas.groupby("ebairrnome")["nescolnome"].nunique().sort_values(ascending=False)
escolas_por_bairro
```

```python
# 4. Descubra o número de alunos aprovados por bairro.
alunos_aprovados_bairro = df_escolas[df_escolas['situ'] == 'AP'].groupby('ebairrnome').size().sort_values(ascending=False)
alunos_aprovados_bairro
```

```python
# 5. Descubra o número de alunos reprovados por frequência que estudam pela MANHÃ em cada escola.
alunos_reprovados = df_escolas[(df_escolas['situ'] == 'RF') & (df_escolas['nturesnome'] == 'MANHÃ')].groupby('nescolnome').size().sort_values(ascending=False)
alunos_reprovados
```

```python
# 6. Descubra o número de alunos reprovados por frequência que estudam a NOITE em cada Bairro

alunos_reprovados_bairro = df_escolas[(df_escolas['situ'] == 'RF') & (df_escolas['nturesnome'] == 'NOITE')].groupby('ebairrnome').size().sort_values(ascending=False)
print(alunos_reprovados_bairro)
```

# Gráfico de Barras

```python
# Carregando csv
link = "http://dados.recife.pe.gov.br/dataset/10a009f2-f9bb-457b-8f78-d8f0dc3ced37/resource/ef521704-6960-4ef1-8f98-a60db4a0d79b/download/pontos_coleta.csv"
df_coletas = pd.read_csv(link, sep=";")
df_coletas.head()
```

```python
# Plotar gráfico de barras que apresentem o número de pontos de coleta por bairro:

x = df_coletas["bairro"].value_counts().index
y = df_coletas["bairro"].value_counts().values

# Criando gráfico
plt.bar(x, y)

plt.title("Contagem de Pontos de Coletas por Bairro")

plt.xlabel("Bairro")
plt.ylabel("Contagem")

plt.show()
```

```python
# Aprimorando o gráfico de barras

plt.figure(figsize=(12,6))
plt.bar(x, y, width = 0.6)

plt.xticks(rotation = 45, ha="right", fontsize = 8)

plt.title("Contagem de Pontos de Coletas por Bairro")

plt.xlabel("Bairro")
plt.ylabel("Contagem")

plt.tight_layout()

plt.show()

```

```python
# Plotar apenas os bairros que tenham 4 ou mais pontos de coleta.
x = df_coletas["bairro"].value_counts()[df_coletas["bairro"].value_counts() >= 4].index
y = df_coletas["bairro"].value_counts()[df_coletas["bairro"].value_counts() >= 4].values

plt.figure(figsize=(12,6))
plt.bar(x, y, width = 0.6)

# Ajuste a inclinação dos rótulos do eixo x e diminua o tamanho da fonte
plt.xticks(rotation=45, ha='right', fontsize=8)

# Adicione um título ao gráfico
plt.title("Contagem de Pontos de Coleta por Bairro")

# Rotule os eixos
plt.xlabel("Bairro")
plt.ylabel("Contagem")

# Mostre o gráfico
plt.tight_layout()  # Ajuste o layout para evitar cortes nos rótulos
plt.show()

```
