# Análise e Visualização de Dados

```python
# Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importar csv
link = "http://dados.recife.pe.gov.br/dataset/10a009f2-f9bb-457b-8f78-d8f0dc3ced37/resource/ef521704-6960-4ef1-8f98-a60db4a0d79b/download/pontos_coleta.csv"
df_coletas = pd.read_csv(link, sep=";")
df_coletas.head()
```

```python
# Verificando se existe alguma coluna com valores NaN
nan_exist = df_coletas.isna().any()

# Imprime as colunas com valores NaN
print("Colunas com valores NaN:")
print(nan_exist[nan_exist])
```

```python
# Substituindo NaN da coluna complemento
df_coletas["complemento"].fillna("s/n", inplace=True)
```

```python
# Entendendo melhor a coluna observação

print(df_coletas["observacao"].sort_values().unique())
```

```python
# Substituindo NaN da coluna observação
df_coletas["observacao"].fillna("Ponto de Coleta", inplace=True)
```

```python
# Verificando se existe alguma coluna com valores NaN
nan_exist = df_coletas.isna().any()

# Imprime as colunas com valores NaN
print("Colunas com valores NaN:")
print(nan_exist[nan_exist])
```

```python
# Entendendo melhor a coluna tiporesiduo
print(df_coletas["tiporesiduo"].sort_values().unique())
```

```python
# Aplicando o replace na coluna tiporesiduo
df_coletas["tiporesiduo"] = df_coletas["tiporesiduo"].str.replace("metal,isopor,papel,pastico,vidro", "metal,isopor,papel,plastico,vidro")
```

```python
print(df_coletas["tiporesiduo"].sort_values().unique())
```

```python
# Entendendo melhor a coluna bairro
print(df_coletas["bairro"].sort_values().unique())
```

```python
# Aplicando o replace na coluna bairro
df_coletas["bairro"] = df_coletas["bairro"].str.replace("Boa viagem", "Boa Viagem")
df_coletas["bairro"] = df_coletas["bairro"].str.replace("Engemho do Meio", "Engenho do Meio")
df_coletas["bairro"] = df_coletas["bairro"].replace("Recife", "Recife Antigo") # Busca pelo termo exato
df_coletas["bairro"] = df_coletas["bairro"].str.replace("Torroes", "Torrões")
df_coletas["bairro"] = df_coletas["bairro"].str.replace(" Novo Carmelo", "Novo Carmelo")
df_coletas["bairro"] = df_coletas["bairro"].str.replace("Poco", "Poço da Panela")
```

```python
# Entendendo melhor a coluna bairro
print(df_coletas["bairro"].sort_values().unique())
```

## Gráfico de barra

```python
# Plotar gráfico de barras que apresentem o número de pontos de coleta por bairro:

x = df_coletas["bairro"].value_counts().index
y = df_coletas["bairro"].value_counts().values

# Criando o gráfico
plt.bar(x, y)

plt.title("Contagem de Pontos de Coletas por Bairro")
plt.xlabel("Bairro")
plt.ylabel("Contagem")

plt.show()
```

```python
# Aprimorando o gráfico

plt.figure(figsize=(12, 6))
plt.bar(x, y, width=0.6)

plt.xticks(rotation=45, ha='right', fontsize=8)

plt.title("Contagem de Pontos de Coletas por Bairro")
plt.xlabel("Bairro")
plt.ylabel("Contagem")

plt.tight_layout()
plt.show()

```

```python
# Plotar apenas os bairros que tenham 4 ou mais pontos de coleta.

bairros_com_muitos_pontos = df_coletas["bairro"].value_counts()[df_coletas["bairro"].value_counts() >= 4]

plt.figure(figsize=(12, 6))
plt.bar(bairros_com_muitos_pontos.index, bairros_com_muitos_pontos.values, width=0.6)

plt.xticks(rotation=45, ha='right', fontsize=8)

plt.title("Contagem de Pontos de Coletas por Bairro")
plt.xlabel("Bairro")
plt.ylabel("Contagem")

plt.tight_layout()
plt.show()
```

## Gráfico de Pizza

```python
# Plotar gráfico de pizza que mostra a contagem e distribuição dos tipos de resíduos que são colhidos.

tipo_residuos = df_coletas["tiporesiduo"].value_counts()

plt.pie(tipo_residuos.values, labels=tipo_residuos.index, autopct='%1.1f%%')

plt.title("Distribuição de Tipos de Resíduos")

plt.show()
```

```python
# Aprimorando o gráfico de pizza

# Criando o explode
tupla_tamanho = (0,) * (df_coletas["tiporesiduo"].value_counts().values.size - 1)
explode = (0.1, ) + tupla_tamanho

plt.figure(figsize=(8,8))

plt.pie(tipo_residuos.values, labels=tipo_residuos.index, autopct='%1.1f%%', explode=explode)

plt.title("Distribuição de Tipos de Resíduos")

plt.axis('equal')

plt.show()
```

```python
# Aprimorando o gráfico de pizza

# Criando o explode
tupla_tamanho = (0,) * (df_coletas["tiporesiduo"].value_counts().values.size - 1)
explode = (0.1, ) + tupla_tamanho

plt.figure(figsize=(8,8))

plt.pie(tipo_residuos.values, autopct='%1.1f%%', explode=explode)

plt.legend(tipo_residuos.index, title="Tipos de Resíduos", loc=(0.9, 0.7))

plt.title("Distribuição de Tipos de Resíduos")

plt.axis('equal')

plt.show()
```

```python
# Exibir um gráfico de pizza que mostra a contagem de distribuição dos
# tipos de resíduos: Óleos de cozinha, pilhas e baterias e eletrônicos.

lista_residuos = ["eletrônico", "pilhas e baterias", "Óleos de Cozinha"]
verifica = df_coletas["tiporesiduo"].isin(lista_residuos)

# Criando o explode
tupla_tamanho = (0,) * (df_coletas[verifica]["tiporesiduo"].value_counts().values.size - 1)
explode = (0.1, ) + tupla_tamanho

plt.figure(figsize=(8,8))

plt.pie(df_coletas[verifica]["tiporesiduo"].value_counts().values, labels=df_coletas[verifica]["tiporesiduo"].value_counts().index, autopct='%1.1f%%', explode=explode, startangle=140)


plt.title("Distribuição de Tipos de Resíduos")

plt.axis('equal')

plt.show()

```
